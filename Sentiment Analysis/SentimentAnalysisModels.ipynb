```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob


# Initialize SpaCy and add the textblob component
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe('spacytextblob')

# Analyze a sample utterance
utterance = "Titanic is a very bad movie. I hate it."
doc = nlp(utterance)
print(doc._.polarity, doc._.subjectivity)
```
